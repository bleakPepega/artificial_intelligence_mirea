```python
import math
import matplotlib.pyplot as plt
import numpy
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
point_1 = numpy.array((0, 0, 0))
point_2 = numpy.array((3, 3, 3))
point_3 = numpy.array((1, 2, 3))
point_4 = numpy.array((0, 2, 2))
ax.scatter(0, 0, 0)
ax.scatter(3, 3, 3)
ax.scatter(1, 2, 3)
ax.scatter(0, 2, 2)
plt.show()
print(math.dist(point_1, point_2), math.dist(point_3, point_4))
```
