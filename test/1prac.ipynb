```python
import math as m
print("Ноль в качестве знака операции"
      "\nзавершит работу программы")
while True:
    s = input("Знак (+,-,*,/, sin, cos, e^(x+y), x^y): ")
    if s == '0':
        break
    if s in ('+', '-', '*', '/', 'sin','cos', 'e^(x+y)', 'x^y'):
        x = float(input("x="))
        y = float(input("y="))
        if s == 'sin':
            print (m.sin(x+y))
        elif s == 'cos':
            print (m.cos(x+y))
        elif s == 'e^(x+y)':
            print (m.exp(x+y))
        elif s == 'x^y':
            print(m.pow(x,y))
        elif s == '+':
            print("%.2f" % (x+y))
        elif s == '-':
            print("%.2f" % (x-y))
        elif s == '*':
            print("%.2f" % (x*y))
        elif s == '/':
            if y != 0:
                print("%.2f" % (x/y))
            else:
                print("Деление на ноль!")
    else:
                print("Неверный знак операции!")
```
